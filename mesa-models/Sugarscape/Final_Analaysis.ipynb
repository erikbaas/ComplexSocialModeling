```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data():
    global amenity, inh_amenity, growth, inh_growth, simple, simple_low_growth
    amenity = pd.read_csv('Data/amenity.csv')
    growth = pd.read_csv('Data/growth.csv')
    inh_growth = pd.read_csv('Data/inh_growth.csv')
    inh_amenity = pd.read_csv('Data/inh_amenity.csv')
    simple = pd.read_csv('Data/simple.csv')
    simple_low_growth = pd.read_csv('Data/simple_low_growth.csv')
    
load_data()
```
