```python
%matplotlib inline
#NetworkX's drawing package is not yet compatible with Python 3

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000
k = 10
ps = np.logspace(-4,0,10)

def c_p_graph(n, k, typ, ps):
    if typ == 'ws':
        gs = [nx.watts_strogatz_graph(n,k,p) for p in ps]
    if typ == 'ba':
        gs = [nx.barabasi_albert_graph(n,k,p) for p in ps]

    clust = np.array([nx.average_clustering(g) for g in gs])
    norm_clust = clust/clust[0]

    path_len = np.array([nx.average_shortest_path_length(g) for g in gs])
    norm_path_len = path_len/path_len[0]

    plt.semilogx(ps, clust)
    plt.semilogx(ps, path_len)
    plt.xlim([.0001,1])
```

```python
c_p_graph(n, k, 'ws', ps)
```

```python
c_p_graph(20, k, 'ba', ps)
c_p_graph(100, k, 'ba', ps)
```
